```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Data = pd.read_csv("TEST.csv")
Data.head()
```

```python
temp_mean = Data["Temperature"].mean()
pres_mean = Data["Pressure"].mean()
hum_mean = Data["Humidity"].mean()
wd_mean = Data["WindDirection(Degrees)"].mean()
```

```python
temp= Data["Temperature"]
pres= Data["Pressure"]
hum= Data["Humidity"]
wd= Data["WindDirection(Degrees)"]
```

```python
parameter = ["Temperature", "Pressure", "Humidity", "WindDirection"]
x_pos = np.arange(len(parameter))
CW = [temp_mean, pres_mean, hum_mean, wd_mean]
```

```python
fig, ax = plt.subplots()
ax.bar(x_pos, CW , align='center', alpha=0.5)
ax.set_ylabel('Average of weather')
ax.set_xticks(x_pos)
ax.set_xticklabels(parameter)
ax.set_title('Weather Report')
ax.yaxis.grid(True)

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(CW, labels=parameter, autopct='%1.1f%%')
ax.axis('equal') # Equal aspect ratio ensures the pie chart is circular.
ax.set_title('Weather Report DataSet')
plt.show()
```

```python
explode = (0.1, 0.4, 0.1, 0.1)
fig, ax = plt.subplots()
ax.pie(CW, explode=explode, labels=parameter, autopct='%1.1f%%', shadow=True, startangle=90)
ax.axis('equal') # Equal aspect ratio ensures the pie chart is circular.
ax.set_title('Weather Report DataSet')
plt.show()
```
